```python
N = 1000
R = 1
u = np.random.rand()

phi = np.random.uniform(0, 2*np.pi, N)
theta = np.random.uniform(0, np.pi, N)
r = R*u**(1/3)


x = r * np.sin(theta) * np.cos(phi)
y = r * np.sin(theta) * np.sin(phi)
z = r * np.cos(theta)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')


ax.scatter(x, y, z, marker='.', color='blue')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')


ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])


plt.show()
```

```python
Ixx = np.average(y**2+z**2)
Iyy = np.average(x**2+z**2)
Izz = np.average(x**2+y**2)
Ixy = np.average (-x*y)
```
