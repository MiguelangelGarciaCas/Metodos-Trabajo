```python
import numpy as np
import matplotlib.pyplot as plt


```

```python
def CreateSphere(N,R=1):
    
    Points = np.zeros((N,3))
        
    for i in (range(N)):
        
        phi = np.random.uniform(0,2*np.pi)
        u = np.random.rand()
        r = R*u**(1/3)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        Points[i] = [r*np.cos(phi)*np.sin(theta),r*np.sin(phi)*np.sin(theta),r*np.cos(theta)]
        
    return Points
```

```python
Sphere=CreateSphere(1000000)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111,projection='3d')
ax.scatter(Sphere[:,0],Sphere[:,1],Sphere[:,2])
```

```python
Ixx=0
Iyy=0
Izz=0
Ixy=0
Ixz=0
Iyz=0
for i in range(Sphere.shape[0]):
    Ixx +=((Sphere[i,1]**2+Sphere[i,2]**2))
    
    Iyy += ((Sphere[i,0]**2+Sphere[i,2]**2))
    Izz += (Sphere[i,0]**2+Sphere[i,1]**2)
    Ixy +=(-Sphere[i,0]*Sphere[i,1])
    Ixz +=(-Sphere[i,0]*Sphere[i,2])
    Iyz+=(-Sphere[i,2]*Sphere[i,2])

```

```python
N=1e6
print(np.average(Ixx)/N)
print(np.average(Iyy)/N)
print(np.average(Izz)/N)
print(np.average(Ixy)/N)
print(np.average(Ixz)/N)
      
print(np.average(Iyz)/N)

```
