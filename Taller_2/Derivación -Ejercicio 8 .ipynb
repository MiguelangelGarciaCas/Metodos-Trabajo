```python
def interpolacion(x, y):
    a0 = y[0]
    a1 = (y[1] - y[0]) / (x[1] - x[0])
    a2 = ((y[2] - y[0]) / (x[2] - x[0]) - a1) / (x[2] - x[1])
    
    p = a0 + a1 * (x - x[0]) + a2 * (x - x[0]) * (x - x[1])
    
    return p
```

```python
import math

def f(x):
    return math.sqrt(math.tan(x))
h = 0.01
x0 = 0.1

df = []
for i in range(10):
    x = x0 + i*h
    aprox = (1/(2*h))*(-3*f(x) + 4*f(x+h) - f(x+2*h))
    df.append(aprox)

print(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sqrt(np.tan(x))

def f_prime(x):
    return (1 / (2 * np.sqrt(np.tan(x)))) * (1 / np.cos(x))**2

a = 0.1
b = 1.1
h = 0.01

x = np.arange(a, b, h)
df_progresiva = (1 / (2 * h)) * (-3*f(x) + 4*f(x+h) - f(x+2*h))
df_central = (1 / (2 * h)) * (f(x+h) - f(x-h))
df_analitica = f_prime(x)

plt.plot(x, df_progresiva, label='Aproximación Progresiva')
plt.plot(x, df_central, label='Aproximación Central')
plt.plot(x, df_analitica, label='Derivada Analítica')
plt.legend()
plt.show()
```

```python
error_progresiva = abs(df_progresiva - df_analitica)
error_central = abs(df_central - df_analitica)

plt.plot(x, error_progresiva, label='Error Aprox. Progresiva')
plt.plot(x, error_central, label='Error Aprox. Central')
plt.legend()
plt.show()
```
